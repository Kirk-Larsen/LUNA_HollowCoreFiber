```julia
using Luna
```

```julia
diameter = 100e-6
a = diameter/2
gas = :Ne
p = 7.0
flength = 1.0

λ0 = 800e-9
τfwhm = 40e-15
energy = 0.100e-3

λlims = (170e-9, 1.6e-6) # grid wavelength limits
trange = 1e-12 # time grid size

modes = 3 # number of modes
```

```julia
first_stage = prop_capillary(a, flength, gas, p; λ0, τfwhm, energy, λlims=λlims, trange=trange, modes)
```

```julia
Plotting.time_1D(first_stage; FTL=true, trange =(-50e-15, 50e-15));
Plotting.spec_1D(first_stage; λrange=(0.17e-6, 1.4e-6), log10=false);
```
