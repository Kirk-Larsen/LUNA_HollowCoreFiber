```julia
using Luna
```

```julia
diameter = 100e-6
a = diameter/2
gas = :Ar
p = 0.5
flength = 1.0

λ0 = 800e-9
τfwhm = 40e-15
energy = 0.110e-3

λlims = (200e-9, 1.4e-6) # grid wavelength limits
trange = 1e-12 # time grid size

modes = 3 # number of modes
```

```julia
first_stage = prop_capillary(a, flength, gas, p; λ0, τfwhm, energy, λlims=λlims, trange=trange, modes)
```

```julia
Plotting.time_1D(first_stage; FTL=true, trange =(-100e-15, 100e-15));
Plotting.spec_1D(first_stage; λrange=(0.20e-6, 1.3e-6), log10=false);
```

```julia
Plotting.prop_2D(first_stage,:λ,trange=(-100e-15, 100e-15),λrange=(0.2e-6, 1.3e-6));
```
